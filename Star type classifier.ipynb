```python
df = pd.read_csv("dataset.csv")
df
```

```python
def one(x):
    return x + 1

def twenty(x):
    return x + 20

df['Luminosity(L/Lo)'] = df['Luminosity(L/Lo)'].apply(one)
df['Luminosity(L/Lo)'] = df['Luminosity(L/Lo)'].apply(np.log)
df['Absolute magnitude(Mv)'] = df['Absolute magnitude(Mv)'].apply(twenty)
```

```python
df1 = pd.get_dummies(df["Spectral Class"], dtype = float)
df2 = pd.get_dummies(df["Star color"], dtype = float)
frames = [df, df1, df2]
frame = pd.concat(frames, axis=1, join="inner")
new_df  = frame.drop(["Spectral Class", "Star color"], axis=1)
new_df[:6]
```

```python
class Network(torch.nn.Module):
    def __init__(self):
        super(Network, self).__init__()
        # 1 слой, принимающий сигнал
        self.fc1=torch.nn.Linear(30, 200)
        # функция сигмоиды для первого слоя
        self.act1=torch.nn.Sigmoid()
        # 2 слой, принимающий сигнал
        self.fc2 = torch.nn.Linear(200, 200)
        # функция сигмоиды для второго слоя
        self.act2 = torch.nn.Sigmoid()
        # 3 функция активации для выходного слоя
        self.fc3=torch.nn.Linear(200, 6)
    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.fc3(x)
        return x
```

```python
our_network = Network() #если слоев > 1?
our_network 
```

```python
optimizer = torch.optim.Adam(our_network.parameters(), 0.01)
criterion = nn.CrossEntropyLoss()
```

```python
for epoch in range(500):
    our_network.train()
    for x_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = our_network(x_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
 
    our_network.eval()
    with torch.no_grad():
        val_outputs =our_network(x_test0)
        val_loss = criterion(val_outputs, y_test0)
#         if val_loss < min_val_loss:
#             min_val_loss = val_loss
#             patience_counter = 0
#         else:
#             patience_counter += 1
 
#         if patience_counter >= early_stopping_patience:
#             print(f"Stopped early at epoch {epoch}")
#             break

our_network.eval()
with torch.no_grad():
    outputs = our_network(x_test0)
    _, predicted = torch.max(outputs, 1)
    accuracy = (predicted == y_test0).sum().item() / len(y_test0)
    print(f'Accuracy: {accuracy}')
```
